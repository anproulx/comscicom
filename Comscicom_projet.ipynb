Les variants du nombre de copies sont des `variations` de la structure chromosomiques de l'ADN par rapport à un `génome de référence`.

```python
# initialiser une liste vide pour base nucléotide
liste_mots_a = []
liste_mots_c = []
liste_mots_g = []
liste_mots_t = []

# faire une liste séparée pour les mots en "A" , "C", "G" et "T"
listes = [liste_mots_a, liste_mots_c, liste_mots_g, liste_mots_t]

for liste,i in zip(listes, range(len(nucleotides))):
  for mot in liste_mots["mots"]:
    if mot[0]== nucleotides[i]:
      liste.append(mot)
```

```python
# imprimer les listes de mots
for liste, nucleotide in zip(listes, nucleotides):
  print("Liste de mots commençant par la lettre {}".format(nucleotide) + "\n" + str(liste) + "\n")
```

```python
def ecrire_poeme(liste_a, liste_t, liste_g, liste_c, regroupements):
  import random
  ''' 
  Écrit le poème suivant les regroupement de bases nucléotides

  Parameters
    ----------
    liste_a : str
        Liste de mots en a
    liste_t : str   
        Liste de mots en t
    liste_g : str
        Liste de mots en g
    liste_c : str
        Liste de mots en c

   Returns
    -------
    poème
  '''
  # définir le seed figer l'état aléatoire
  np.random.seed(seed=123)
  
  lettre = ["A", "T", "G", "C"]
  poeme = ""

  # pour chaque vers
  for regr in regroupements:

    # initier un vers vide
    vers = ""

    # pour chaque lettre
    for i in range(len(regr)):

      # choisir un mot aléatoire selon 
      # la liste de mots correspondante
      
      if regr[i] == "A":
        vers += random.choice((liste_a)) + " "

      elif regr[i] == "T":
        vers += random.choice((liste_t)) + " "
        
      elif regr[i] == "G":
        vers += random.choice((liste_g)) + " "

      elif regr[i] == "C":
        vers += random.choice((liste_c)) + " "

      else :
        vers += ","

    poeme += vers + "\n"

  return print(poeme + "\n")
```

```python
ecrire_poeme(liste_mots_a, liste_mots_t, liste_mots_g, liste_mots_c, regroupements)
```
